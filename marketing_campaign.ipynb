## 4. Apply necessary transformations for the feature variables

We are using yeo-johnson transformation for Income and CCAvg columns

```python
pt = PowerTransformer(standardize=False)
pt.fit(data.Income.values.reshape(-1,1))
temp = pt.transform(data.Income.values.reshape(-1,1))
data['Income_t'] = temp
sns.distplot(temp)
```

```python
pt = PowerTransformer(standardize=False)
pt.fit(data.CCAvg.values.reshape(-1,1))
temp = pt.transform(data.CCAvg.values.reshape(-1,1))
data['CCAvg_t'] = temp
sns.distplot(temp)
```

- Observation: As we can observe that these are now looking more normal, so now we have to drop our old Income and CCAvg columns .

```python
data.drop(['Income', 'CCAvg'], axis=1, inplace=True)
```

- We can not use yeo-johnson transformation here because we have 347 different values here and most of them are towards zero, so we are using Binning here.

```python
data['Mortgage_int'] = pd.cut(data.Mortgage, bins= [0,100,200,300,400,500,600,700], labels= [0,1,2,3,4,5,6],include_lowest=True)
data.drop('Mortgage', axis =1, inplace=True)
```

## 5. Split you dataset into Train Test Split

- It's time to split our dataset into training and testing sets
- First we divide our dataframe into X and Y

```python
X = data.drop('Personal Loan', axis=1)
y = data[['Personal Loan']]
```

```python
#Shape of X and y
print(f'X = {X.shape}')
print(f'y = {y.shape}')
```

- Splitting the dataset into 70:30 ratio for training and testing, also using stratify in order to maintain the percentage of true values.

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y, random_state=1)
```

```python
print(f'X_train = {X_train.shape}')
print(f'y_train = {y_train.shape}')
print(f'X_test = {X_test.shape}')
print(f'y_test = {y_test.shape}')
```

## 6. Prediction Models and Comparisons

### ● Logistic Regression Model

Logistic regression is a statistical model that in its basic form uses a logistic function to model a binary dependent variable, although many more complex extensions exist

```python
LR_model = LogisticRegression(fit_intercept=True, random_state=1, max_iter=250)
LR_model.fit(X_train, np.array(y_train).ravel())
LR_predict = LR_model.predict(X_test)
LR_train_acc = LR_model.score(X_train,y_train)
LR_test_acc = LR_model.score(X_test,y_test)
```

```python
print(f'LogisticRegression Train Accuracy = {LR_train_acc}')
print(f'LogisticRegression Test Accuracy = {LR_test_acc}\n\n')
print(classification_report(y_test, LR_predict, target_names=['0','1']))
print(f'ROC AUC Score = {roc_auc_score(y_test, LR_predict)}')
```

```python
sns.heatmap(confusion_matrix(y_test, LR_predict), fmt='0.2f', annot=True , cmap='plasma')
plt.ylabel('Observed')
plt.xlabel('Predicted')
plt.show()
```

### ● Decision Tree Model

Decision tree learning is one of the predictive modelling approaches used in statistics, data mining and machine learning. It uses a decision tree (as a predictive model) to go from observations about an item (represented in the branches) to conclusions about the item's target value (represented in the leaves). Tree models where the target variable can take a discrete set of values are called classification trees, we are using classification tree here.

```python
dt_model = DecisionTreeClassifier(criterion='entropy', max_depth=6,random_state=1)
dt_model.fit(X_train, y_train)
dt_predict = dt_model.predict(X_test)
dt_train_acc = dt_model.score(X_train,y_train)
dt_test_acc = dt_model.score(X_test,y_test)
```

```python
print(f'DecisionTreeClassifier Train Accuracy = {dt_train_acc}')
print(f'DecisionTreeClassifier Test Accuracy = {dt_test_acc}\n\n')
print(classification_report(y_test, dt_predict, target_names=['0','1']))
print(f'ROC AUC Score = {roc_auc_score(y_test, dt_predict)}')
```

```python
sns.heatmap(confusion_matrix(y_test, dt_predict), fmt='0.2f', annot=True , cmap='plasma')
plt.ylabel('Observed')
plt.xlabel('Predicted')
plt.show()
```

### ● Random Forest Model

Random forests or random decision forests are an ensemble learning method for classification, regression and other tasks that operate by constructing a multitude of decision trees at training time and outputting the class that is the mode of the classes or mean prediction of the individual trees.

```python
rf_model = RandomForestClassifier(criterion='entropy', max_depth=9,random_state=1)
rf_model.fit(X_train, np.array(y_train).ravel())
rf_predict = rf_model.predict(X_test)
rf_train_acc = rf_model.score(X_train,y_train)
rf_test_acc = rf_model.score(X_test,y_test)
```

```python
print(f'RandomForestClassifier Train Accuracy = {rf_train_acc}')
print(f'RandomForestClassifier Test Accuracy = {rf_test_acc}\n\n')
print(classification_report(y_test, rf_predict, target_names=['0','1']))
print(f'ROC AUC Score = {roc_auc_score(y_test, rf_predict)}')
```

```python
sns.heatmap(confusion_matrix(y_test, rf_predict), fmt='0.2f', annot=True , cmap='plasma')
plt.ylabel('Observed')
plt.xlabel('Predicted')
plt.show()
```

### ● Naive Bayes

In statistics, Naive Bayes classifiers are a family of simple "probabilistic classifiers" based on applying Bayes' theorem with strong independence assumptions between the features. They are among the simplest Bayesian network models

```python
naive_model = GaussianNB()
naive_model.fit(X_train, np.array(y_train).ravel())
naive_predict = naive_model.predict(X_test)
naive_train_acc = naive_model.score(X_train,y_train)
naive_test_acc = naive_model.score(X_test,y_test)
```

```python
print(f'Naive Bayes Train Accuracy = {naive_train_acc}')
print(f'Naive Bayes Test Accuracy = {naive_test_acc}\n\n')
print(classification_report(y_test, naive_predict, target_names=['0','1']))
print(f'ROC AUC Score = {roc_auc_score(y_test, naive_predict)}')
```

```python
sns.heatmap(confusion_matrix(y_test, naive_predict), fmt='0.2f', annot=True , cmap='plasma')
plt.ylabel('Observed')
plt.xlabel('Predicted')
plt.show()
```

## ● KNN

The k-nearest neighbors algorithm (k-NN) is a non-parametric method proposed by Thomas Cover used for classification and regression. In both cases, the input consists of the k closest training examples in the feature space. The output depends on whether k-NN is used for classification or regression, we are using it here for classification.

```python
knn_model = KNeighborsClassifier(n_neighbors= 15)
knn_model.fit(X_train, np.array(y_train).ravel())    
knn_predict = knn_model.predict(X_test)
knn_train_acc = knn_model.score(X_train,y_train)
knn_test_acc = knn_model.score(X_test,y_test)
```

```python
print(f'KNN Train Accuracy = {knn_train_acc}')
print(f'KNN Test Accuracy = {knn_test_acc}\n\n')
print(classification_report(y_test, knn_predict, target_names=['0','1']))
print(f'ROC AUC Score = {roc_auc_score(y_test, knn_predict)}')
```

```python
sns.heatmap(confusion_matrix(y_test, knn_predict), fmt='0.2f', annot=True , cmap='plasma')
plt.ylabel('Observed')
plt.xlabel('Predicted')
plt.show()
```
